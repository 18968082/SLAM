```python
import numpy as np
```

```python
def vector_to_homogen(vector): # [ x y z alpha beta gamma]
    ca = np.cos(vector[3]) 
    sa = np.sin(vector[3])
    cb = np.cos(vector[4]) 
    sb = np.sin(vector[4])
    cg = np.cos(vector[5]) 
    sg = np.sin(vector[5])
    tx, ty, tz = vector[0], vector[1],  vector[2]
    vector_homogen = np.array([[ ca*cb , ca*sb*sg - sa*cg, ca*sb*cg + sa*sg, tx ],
                               [ sa*cb , sa*sb*sg + ca*cg, sa*sb*cg - ca*sg, ty ],
                               [ -sb   , cb*sg           , cb*sg           , tz ],
                               [ 0     , 0               , 0               , 1  ]])
    return vector_homogen
```

```python
transformation = np.array([-0.09174403, # x
                           -0.12822234, # y
                            0.00436085, # z
                            0.00040659, # alpha
                           -0.00028076, # beta
                            0.00157765])# gamma

transformation_homo = vector_to_homogen(transformation)
```
